```python
import pandas as pd


file = "Sankey_example.csv"

#Converting data before starting analysis
df = pd.read_csv(file)
```

```python
def PrepToSankey(bd,Origem,Destino,Peso):
    # bd = Pandas DataFrame to be analyzed
    # Origem = Source column in bd
    # Destino = Destination column in bd
    # Peso = Weight column in bd
    
    x1 = bd[Origem].unique()
    x1 = x1[~pd.isnull(x1)]
    x1 = list(x1)
    xQ1 = list(bd[Origem])

    x2 = bd[Destino].unique()
    x2 = x2[~pd.isnull(x2)]
    x2 = list(x2)
    xQ2 = list(bd[Destino])

    A = []
    B = []
    C = []

    for i in range (0,len(x1)):
        for j in range (0,len(x2)):
            A.append(x1[i])    
            B.append(x2[j])
            C.append(bd[(bd[Origem] == x1[i]) & (bd[Destino] == x2[j])][Peso].sum())    

    a = pd.DataFrame()
    a[Origem] = A
    a[Destino]= B
    a[Peso]= C
    df_sankey = a[a[Peso] != 0]
            
    return df_sankey 
```

```python

data_sankey_diag = PrepToSankey(df,'Vendedor','Cliente','Valor')
```
